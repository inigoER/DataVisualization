```python
dj = sns.load_dataset("dowjones")
dj
```

### 1. Utiliza Plotly express para hacer un gráfico de línea con los datos de 1965

No olvides agregar un buen título.

```python
dj65 = dj[dj['Date'].dt.year == 1965]
dj65
```

```python
fig = px.line(dj65, x="Date", y="Price", title='Dowjones price across the year 1965')
fig.show()
```

### 2. Modifica el parámetro _text_ para que cada punto tenga su mes indicado.

```python
fig = px.line(dj65, x="Date", y="Price", title='Dowjones price across the year 1965', text = dj65['Date'].dt.month_name())
fig.show()
```

### 3. Repite el gráfico anterior, pero agrega _markers_

```python
fig = px.line(dj65, x="Date", y="Price", title='Dowjones price across the year 1965', text = dj65['Date'].dt.month_name(), markers = True)
fig.show()
```

## 5. Crea un gráfico de barras con los precios de cada año de la década de los 60, usando Plotly Express.

Puedes utilizar el precio inicial, el final, el promedio, o algun otro que se te ocurra por cada año.

```python
years = []
averageprice = []
dj60.reset_index(drop = True, inplace = True)
for i in dj60['Date'].dt.year.unique():
  years.append(i)
  avg = 0
  for j in range(0, len(dj60['Date'])):
    if dj60['Date'][j].year == i:
      avg = avg + dj60['Price'][j]
  avg = avg/12
  averageprice.append(avg)
years_pd = pd.DataFrame(years)
years_pd.columns = ['Year']
averageprice_pd = pd.DataFrame(averageprice)
averageprice_pd.columns = ['Average Price']
djDecade = years_pd.join(averageprice_pd)
djDecade
```

```python
fig = px.bar(djDecade, x='Year', y="Average Price", title="Average price of Dowjones per year")
fig.show()
```
